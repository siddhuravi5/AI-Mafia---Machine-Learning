```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#step 1 is to get the training data
dfx=pd.read_csv('Linear_X_Train.csv')
dfy=pd.read_csv('Linear_Y_Train.csv')
```

```python
x=dfx.values
x=x.reshape(-1)
x.shape
#x=x[1:2500]
x.shape
```

```python
y=dfy.values
y=y.reshape(-1)
y.shape
#y=y[1:2500]
y.shape
```

```python
x=(x-x.mean())/x.std()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
def hypothesis(x,theta):
    return (theta[0]+theta[1]*x)
```

```python
def error(x,y,theta):
    err=0
    for i in range(x.shape[0]):
        hx=hypothesis(x[i],theta)
        err+=(hx-y[i])**2
    return err
```

```python
def gradient(x,y,theta):
    grad=np.zeros((2,))
    for i in range(x.shape[0]):
        hx=hypothesis(x[i],theta)
        grad[0]+=(hx-y[i])
        grad[1]+=((hx-y[i])*x[i])
    return grad
    
```

```python
def gradient_descent(x,y,learning_rate=0.00005):
    #assume or provide some random value to theta to start with
    theta=np.array([0.0,1.0])
    itr=0
    final_itr=1000
    error_list=[]
    theta_list=[]
    
    while(itr<=final_itr):
        grad=gradient(x,y,theta)
        e=error(x,y,theta)
        error_list.append(e)
        theta_list.append((theta[0],theta[1]))
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
        itr+=1
    return theta,error_list,theta_list

        
```

```python
final_theta,error_list,theta_list=gradient_descent(x,y)
```

```python
plt.plot(error_list)
plt.show()
```
